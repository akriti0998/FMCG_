```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Raw Sales Data.csv')
df.head()
```

```python
df['Qty']=df['Qty'].str.replace(',','')
df['Sales']=df['Sales'].str.replace(',','')
```

```python
df.rename(columns={'Sub-Channels':'Sub_Channels'},inplace=True)
df.columns
```

```python
df['Month']= pd.to_datetime(df['Month'])
df[['Qty', 'Sales']] = df[['Qty', 'Sales']].astype(int)
```

```python
df.info()
```

```python
#Question 1)Monthwise number of quantities sold
df_month=df.groupby(df.Month)['Qty'].sum()
df_month
```

```python
#Question 2) Category wise Qty sold
df_Category=df.groupby(df.Category)['Qty'].sum()
df_Category
```

```python
#Question 3) Sub-channel wise total sales
df_sub_chan=df.groupby(df.Sub_Channels)['Sales'].sum()
df_sub_chan
```

```python
#Question 4) Highest sale in which month
df_high_sale=df[["Month","Sales"]][df['Sales']==df["Sales"].max()]
df_high_sale
```

```python
#Question 5) Highest selling product (based on sales) (packaging quantity doesn't matter Ex. consider "GET SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)
df["base_product_name"] = df["Product Name"].str.replace('\d+','1')
df["base_product_name"] = df["base_product_name"].apply(lambda x:x.split('1') [0])
most_sell_product = pd.DataFrame(df.groupby("base_product_name")["Qty"].sum().reset_index())
most_sell_product
```

```python
 #Question 6)List of different products company is offering (packaging quantity doesn't matter. Ex. consider "GET SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)
product_list = pd.DataFrame(df["base_product_name"].unique())
product_list
```
